```python
x=df.drop(labels=['TARGET'],axis=1)
```

```python
y=df['TARGET']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
from sklearn.model_selection import train_test_split
X_train.shape,X_test.shape
```

Lets apply the variance threshold 

```python
var_thrsh=VarianceThreshold(threshold=0)
var_thrsh.fit(x)
```

Finding the non constant features

```python
var_thrsh.get_support()
```

```python
sum(var_thrsh.get_support())
```

```python
constant_columns = [column for column in x.columns
                    if column not in x.columns[var_thrsh.get_support()]]

print(len(constant_columns)) 
```

```python
for column in constant_columns:
    print(column)
```

lets drop the columns which are not useful for EDA
